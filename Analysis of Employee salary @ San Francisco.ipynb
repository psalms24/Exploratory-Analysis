#### Counting the number of missing rows in each column

```python
missing_value_count = dataset.isnull().sum(axis=0)
print(missing_value_count)
```

#### Finding the total missing value and the percentage missing

```python
total_cell = np.product(dataset.shape)
total_missing_value = missing_value_count.sum()
print("The total missing value is:", total_missing_value)

percent_missing = (total_missing_value/total_cell) * 100
print('The percentage missing is:', percent_missing)
```

#### Remove all columns with at least one missing value

```python
column_with_na_value = dataset.dropna(axis ="columns")
column_with_na_value.head() #We could also use this command dataset.dropna(axis=1), row =0, column=1
```

#### Checking how many columns did we lose

```python
total_columns_originally = int(dataset.shape[1])
total_columns_retained = int(column_with_na_value.shape[1])
Columns_lost = total_columns_originally - total_columns_retained

print ("The original columns are:", total_columns_originally)

print ("The retained columns are:", total_columns_retained)

print("The lost columns are:", Columns_lost)
```
