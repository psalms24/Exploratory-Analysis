```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
df = pd.DataFrame()
sales_name = ['Ahmed', 'Omar', 'Ali', 'Ziad', 'Salwa', 'Lila']

mobile_sales = [2540, 1370, 1320, 2000, 2100, 2150]

Tv_sales = [2200, 1900, 2150, 1850, 1770, 2000]

df ['Name']= sales_name 
df ['Mobile_Sales'] = mobile_sales
df ['Tv_Sales'] = Tv_sales
df.set_index('Name', drop=True, inplace=True)

df
```

```python
df.plot.bar(figsize=(10, 5), rot=0).legend(bbox_to_anchor=(1.1, 1))
plt.xlabel('Salesmen')
plt.ylabel('Sales')
plt.title('Sales volume for two salesmen')
plt.show()
```

```python
df.plot.pie(figsize=(10, 5),subplots=True)
```

```python
df.plot.box(figsize=(10, 5))
```

```python
df.plot.area(figsize=(10,5)).legend(bbox_to_anchor=(1.2,1))
```

```python
df.plot.bar(stacked=True, figsize=(10,5)).legend(bbox_to_anchor=(1.3,1))
```

```python
df2= pd.DataFrame()

Name = ['Ahmed', 'Omar', 'Ali', 'Ziad', 'Salwa', 'Lila']

Rice_sales = [2900, 3000, 2100, 3200, 2550, 4000]

df2['Name']= Name
df2['Rice_Sales']= Rice_sales

df2.set_index('Name', inplace=True)

df2
```

```python
df3=pd.merge(df, df2, on= 'Name')
df3
```

```python
df3.plot.bar(figsize=(10, 5), rot=0).legend(bbox_to_anchor=(1.1, 1))
plt.xlabel('Salesmen')
plt.ylabel('Sales')
plt.title('Sales volume for two salesmen')
plt.show()
```
